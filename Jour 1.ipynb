```python
import yfinance as yf
import numpy as np
import pandas as pd

# Fonction principale
def fetch_and_analyze_stock(ticker_symbol, period='1mo'):
    try:
        # Téléchargement des données
        stock_data = yf.download(ticker_symbol, period=period, interval='1d')
        
        # Vérification que des données sont disponibles
        if stock_data.empty:
            print(f"Aucune donnée trouvée pour le ticker {ticker_symbol}.")
            return
        
        # Afficher un aperçu des données
        print("Aperçu des données :")
        print(stock_data.head())
        
        # Extraction des prix de clôture
        closing_prices = stock_data['Close']
        
        # Calcul des statistiques
        average_price = np.mean(closing_prices)
        std_dev_price = np.std(closing_prices)
        max_price = np.max(closing_prices)
        min_price = np.min(closing_prices)
        
        # Affichage des statistiques
        print("\nStatistiques :")
        print(f"Moyenne : {average_price:.2f}")
        print(f"Écart type : {std_dev_price:.2f}")
        print(f"Maximum : {max_price:.2f}")
        print(f"Minimum : {min_price:.2f}")
    
    except Exception as e:
        print(f"Erreur : {e}")
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import streamlit as st

# Télécharger les données
def fetch_stock_data(ticker_symbol, period='1mo'):
    stock_data = yf.download(ticker_symbol, period=period, interval='1d')
    return stock_data

# Créer le dashboard
def main():
    st.title("Dashboard des Actions")
    st.sidebar.header("Paramètres")
    
    # Sélection de l'action et de la période
    ticker = st.sidebar.text_input("Entrez le ticker de l'action :", value="KO")  # Exemple : Coca-Cola
    period = st.sidebar.selectbox("Sélectionnez la période :", ["1mo", "3mo", "6mo", "1y", "2y", "5y"])
    
    # Télécharger les données
    stock_data = fetch_stock_data(ticker, period)
    
    if stock_data.empty:
        st.write("Aucune donnée disponible pour ce ticker.")
        return
    
    st.write(f"### Données pour {ticker} ({period})")
    st.dataframe(stock_data.head())
    
    # Calcul des statistiques
    closing_prices = stock_data['Close'].dropna()
    average_price = closing_prices.mean()
    std_dev_price = closing_prices.std()
    max_price = closing_prices.max()
    min_price = closing_prices.min()
    
    # Afficher les statistiques
    st.write("### Statistiques")
    st.write(f"Moyenne : **{average_price:.2f}**")
    st.write(f"Écart type : **{std_dev_price:.2f}**")
    st.write(f"Maximum : **{max_price:.2f}**")
    st.write(f"Minimum : **{min_price:.2f}**")
    
    # Graphique 1 : Évolution des prix de clôture
    st.write("### Graphique 1 : Évolution des Prix de Clôture")
    st.line_chart(closing_prices)
    
    # Graphique 2 : Volume des transactions
    st.write("### Graphique 2 : Volume des Transactions")
    st.bar_chart(stock_data['Volume'])
    
    # Graphique 3 : Distribution des prix de clôture
    st.write("### Graphique 3 : Distribution des Prix de Clôture")
    fig, ax = plt.subplots(figsize=(10, 5))
    ax.hist(closing_prices, bins=20, color='blue', edgecolor='black', alpha=0.7)
    ax.set_title("Distribution des Prix de Clôture")
    ax.set_xlabel("Prix de Clôture")
    ax.set_ylabel("Fréquence")
    st.pyplot(fig)

# Lancer l'application
if __name__ == "__main__":
    main()
```

```python
import yfinance as yf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Télécharger les données
def fetch_stock_data(ticker_symbol, period='1mo'):
    stock_data = yf.download(ticker_symbol, period=period, interval='1d')
    return stock_data

# Fonction pour afficher une heatmap
def display_heatmap(ticker, period='1mo'):
    # Télécharger les données
    stock_data = fetch_stock_data(ticker, period)
    
    if stock_data.empty:
        print(f"Aucune donnée disponible pour le ticker {ticker}.")
        return
    
    # Sélection des colonnes pertinentes
    relevant_data = stock_data[['Open', 'High', 'Low', 'Close', 'Volume']]
    
    # Calcul de la matrice de corrélation
    correlation_matrix = relevant_data.corr()
    
    # Création de la heatmap avec Seaborn
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True, square=True)
    plt.title(f"Heatmap des Corrélations pour {ticker}")
    plt.show()

# Exemple d'utilisation
ticker = "KO"  # Remplacez par le ticker de l'action souhaitée
display_heatmap(ticker, period='1mo')
```
