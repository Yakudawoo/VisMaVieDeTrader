```python
import yfinance as yf
from ta.momentum import rsi
import pandas as pd

# Fonction pour calculer le RSI
def calculate_rsi(data):
    # Vérifier que la colonne 'Close' est disponible
    if 'Close' not in data.columns:
        print("Données manquantes : La colonne 'Close' est requise pour calculer le RSI.")
        return None
    
    # Supprimer les valeurs manquantes dans 'Close'
    data = data.dropna(subset=['Close'])
    
    # Calcul du RSI (fenêtre de 14 jours)
    data['RSI'] = rsi(data['Close'], window=14)
    return data

# Fonction pour analyser plusieurs actions
def analyze_stocks(tickers, period='6mo'):
    results = {}
    for ticker in tickers:
        print(f"Téléchargement des données pour {ticker}...")
        stock_data = yf.download(ticker, period=period, interval='1d')
        
        # Vérifier si les données sont disponibles
        if stock_data.empty:
            print(f"Aucune donnée disponible pour {ticker}.")
            continue
        
        # Calculer le RSI
        processed_data = calculate_rsi(stock_data)
        if processed_data is not None:
            results[ticker] = processed_data[['Close', 'RSI']]  # Conserver uniquement 'Close' et 'RSI'
        else:
            print(f"RSI non calculé pour {ticker} (colonnes manquantes ou données invalides).")
    
    return results

```
